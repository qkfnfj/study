```python
import pandas as pd
import numpy as np
series = pd.read_csv('환율27.csv')
```

```python
from matplotlib import pyplot as plt
```

```python
series.dtypes
```

```python
series = series.astype({'date':'str'})
```

```python
series.dtypes
```

```python
series
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(series['price'])
```

```python
plot_pacf(series['price'])
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(series['price'], order=(0,1,1))
model_fit = model.fit(trend='c',full_output=True, disp=1)
print(model_fit.summary())
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(series['price'], order=(0,1,1))
model_fit = model.fit(trend='nc',full_output=True, disp=1)
print(model_fit.summary())
```

```python
model_fit.plot_predict()
```

```python
diff_1=series['price'].diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
diff_1=series['price'].diff(periods=1).iloc[5:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```
