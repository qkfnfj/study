```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
%matplotlib inline
n_samples = 30
X = np.sort(np.random.rand(n_samples))
y = np.sin(2*np.pi*X) + np.random.randn(n_samples)*0.1
X = X.reshape(X.shape[0], 1)
print(X.shape, y.shape)
```

```python
font_size = 16
plt.figure(figsize=(10, 5))
plt.scatter(x=X, y=y, c='blue', marker='o', label='sample') # 파란색 동그라미로 시각화
plt.xlabel('X-axis', fontsize=font_size) # x축의 이름과 글자 크기
plt.ylabel('Y-axis', fontsize=font_size) # y축의 이름과 글자 크기
plt.xticks(fontsize=font_size) # x축의 숫자 크기
plt.yticks(fontsize=font_size) # y축의 숫자 크기
plt.title('Plot of sample', fontsize=(font_size+2))
plt.legend(fontsize=font_size)
plt.show()
```

```python
ordinary_regressor = LinearRegression().fit(X, y) # 선형회귀분석
pred_ordinary = ordinary_regressor.predict(X) # 예측
b1_ordinary = ordinary_regressor.coef_
print("Coef of ordinary regressor: ", b1_ordinary)
ridge_regressor = Ridge(alpha=0.25).fit(X, y) # Ridge 회귀분석
pred_ridge = ridge_regressor.predict(X) # 예측
b1_ridge = ridge_regressor.coef_
print("Coef of ridge regressor: ", b1_ridge)
```

```python
print("회귀계수가 줄어들었나요? 맞으면 True, 틀리면 False => ",
abs(b1_ordinary) > abs(b1_ridge))
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x=X, y=y, c='blue', marker='o', label='sample')
plt.plot(X, pred_ordinary, c='red', linestyle='--', label='pred(ordinary)')
plt.plot(X, pred_ridge, c='green', linestyle='-.', label='pred(ridge)')
plt.xlabel('X-axis', fontsize=font_size)
plt.ylabel('Y-axis', fontsize=font_size)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.legend(fontsize=font_size)
plt.show()
```

```python
var_pred_ordinary = np.var(pred_ordinary) # ordinary 회귀 기반 예측값의 분산
var_pred_ridge = np.var(pred_ridge) # ridge 회귀 기반 예측값의 분산
print("분산이 줄어들었나요? 맞으면 True, 틀리면 False => ",
var_pred_ordinary > var_pred_ridge)
```

```python
plt.figure(figsize=(10, 5))
sample_x = [range(n_samples)]
plt.scatter(x=sample_x, y=pred_ordinary, c='red', marker='s', s=70, label='pred(ordinary)')
plt.scatter(x=sample_x, y=pred_ridge, c='green', marker='^', s=80, label='pred(ridge)')
plt.xlabel('Samples', fontsize=font_size)
plt.ylabel('Predicted value', fontsize=font_size)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.legend(fontsize=font_size)
plt.show()
```

```python
range_t = [0, 0.01, 0.1, 1, 2, 4, 6, 8, 10]
ridge_coef = []
for t in range_t:
  ridge_regressor = Ridge(alpha=t).fit(X, y)
  ridge_coef.append(ridge_regressor.coef_)
plt.figure(figsize=(10, 5))
plt.scatter(x=range_t, y=ridge_coef, label='b1_ridge')
plt.axhline(b1_ordinary, c='red', label='b1_ordinary')
plt.xlabel('t', fontsize=font_size)
plt.ylabel('coefficient', fontsize=font_size)
plt.legend(fontsize=font_size)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import Lasso
%matplotlib inline
boston = load_boston()
X = boston.data
X = X[:, 0:3]
y = boston.target
print(X.shape, y.shape)
```

```python
range_t = [0, 0.01, 0.1, 1, 2, 4, 6, 8, 10, 30, 50, 70, 90, 100]
lasso_coef = []
for t in range_t:
 lasso_regressor = Lasso(alpha=t).fit(X, y)
 lasso_coef.append(lasso_regressor.coef_)
coef_df = pd.DataFrame(lasso_coef, columns=['crim', 'zn', 'indus'])
print(coef_df.shape)
coef_crim = coef_df['crim'].tolist()
coef_zn = coef_df['zn'].tolist()
coef_indus = coef_df['indus'].tolist()
```

```python
font_size = 16
plt.figure(figsize=(10, 5))
plt.plot((range_t), coef_crim, c='red', linestyle='-', label='CRIM')
plt.plot((range_t), coef_zn, c='blue', linestyle='--', label='ZN')
plt.plot((range_t), coef_indus, c='green', linestyle='-.', label='INDUS')
plt.axhline(0, c='gray', linestyle=':')
plt.xlabel('t', fontsize=font_size)
plt.ylabel('Coefficient', fontsize=font_size)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.legend(fontsize=font_size)
plt.show()
```
