```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
import matplotlib.pyplot as plt
plt.scatter(diabetes.data[:, 2], diabetes.target) # 입력데이터의 3번째 변수와 타깃데이터
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = diabetes.data[:, 2] # 3번째 변수만 선택
y = diabetes.target
```

```python
w = 1.0
b = 1.0
```

```python
y_hat = x[0] * w + b
```

```python
w_inc = w + 0.1
y_hat_inc = w_inc * x[0] + b
print(y_hat_inc) # y-hat_inc가 y_hat(1.0616962065186886 )보다 조금 증가
```

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

```python
w_new = w + w_rate
print(w_new)
```

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
b_new = b + 1
print(b_new)
```

```python
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

```python
for x_i, y_i in zip(x, y):
 y_hat = x_i * w + b
 err = y_i - y_hat
 w_rate = x_i
 w = w + w_rate * err
 b = b + 1 * err
print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_hat = x[2] * w_new + b_new
err = y[2] - y_hat
w_rate = x[2]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

```python
for i in range(1, 100):
 for x_i, y_i in zip(x, y):
  y_hat = x_i * w + b
  err = y_i - y_hat
  w_rate = x_i
  w = w + w_rate * err
  b = b + 1 * err
print(w, b)
```
